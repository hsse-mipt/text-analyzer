```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv('../../../static/datasets/original/train_data.csv', sep='\t')
```

```python
sent_label = ['sentence', 'label']
NER_label = ['entity_tag', 'label']
sent_label_data = train_data[sent_label]
NER_data = train_data[NER_label]
#sent_label_data['label'].value_counts()
```

```python
fig, ax = plt.subplots()

labels = ['Нейтральные', 'Негативные', 'Позитивные']
counts = sent_label_data['label'].value_counts().tolist()
bar_colors = ['tab:grey', 'tab:red', 'tab:green']

ax.bar(labels, counts, color=bar_colors)

ax.set_ylabel('Количество предложений')
ax.set_title('Распределение тональностей')

plt.show()
```

```python
NER_data.head()
NERs = ['PERSON','PROFESSION','ORGANIZATION','COUNTRY','NATIONALITY']
NER_count = NER_data['entity_tag'].value_counts().tolist()
```

```python
fig, ax = plt.subplots()

ax.bar(NERs, NER_count)

ax.set_ylabel('Кол-во появлений')
ax.set_title('Распределение NER-сущностей')

plt.show()
```

```python
NER_count_pos = NER_data[NER_data['label'] == 1]['entity_tag'].value_counts().tolist()
fig, ax = plt.subplots()

ax.bar(NERs, NER_count_pos, color="tab:green")

ax.set_ylabel('Кол-во появлений')
ax.set_title('Распределение NER-сущностей в положительных предложениях')

plt.show()
```

```python
NER_count_neu = NER_data[NER_data['label'] == 0]['entity_tag'].value_counts().tolist()
fig, ax = plt.subplots()

ax.bar(NERs, NER_count_neu, color="tab:grey")

ax.set_ylabel('Кол-во появлений')
ax.set_title('Распределение NER-сущностей в нейтральных предложениях')

plt.show()
```

```python
NER_count_neg = NER_data[NER_data['label'] == -1]['entity_tag'].value_counts().tolist()
fig, ax = plt.subplots()

ax.bar(NERs, NER_count_neg, color="tab:red")

ax.set_ylabel('Кол-во появлений')
ax.set_title('Распределение NER-сущностей в негативных предложениях')

plt.show()
```

```python
sentences = dict()
for s in train_data.sentence:
    if len(s) not in sentences.keys():
        sentences[len(s)] = 0
    sentences[len(s)] += 1

lengths = list(sentences.keys())
grid = list(sentences.values())
```

```python
plt.figure()
plt.hist(grid, density=True)
plt.ylabel('Относительное появление')
plt.xlabel('Длина предложений')
plt.title('Распределение длин предложений')
plt.show()
```
